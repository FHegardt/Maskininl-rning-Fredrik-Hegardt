```python
#0 Iris flower dataset
#a)
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = datasets.load_iris(as_frame = True)

```

```python

df = df.frame
df.head(15)
```

```python
first_list= ["sepal width (cm)", "petal width (cm)"]
second_list = ["sepal length (cm)", "petal length (cm)" ]
for a in range(2):
    for b in range(2):
        fig, ax = plt.subplots(dpi=80)
        sns.scatterplot(data = df, x =first_list[a], y =second_list[b], hue="target")
```
