```python
from utils import plot_svm_margins
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

blobs = make_blobs([300, 300,300], 2, centers = [(-2,-2), (2,2), (6,6)], random_state=42)[0]
plt.scatter(blobs[:,0], blobs[:,1])
```

## Clustering
- K-means algorithm
- unsupervised learning algorithm

```python
from sklearn.cluster import KMeans

kmean = KMeans(3)
kmean.fit(blobs)
blobs = np.c_[blobs, kmean.predict(blobs)]
blobs
```

```python
df_blobs = pd.DataFrame(blobs, columns = ["X1", "X2","X3", "label"])

df_blobs.loc[df_blobs["label"] == 1 , "label"] = 0
df_blobs


sns.scatterplot(data=df_blobs, x="X1", y="X2", hue = "label")
```

```python
X, y = df_blobs.drop("label", axis=1).to_numpy(), df_blobs["label"].to_numpy()
X, y
```

## SVM
- S = Support
- V = Vector
- M = Machine

### Linerar kernal

```python
fig, ax = plt.subplots(1, 3, dpi=100, figsize = (12, 4))

for i, C in enumerate([10, .1, .01], 1):
    plt.subplot(1,3,i)
    classifier = SVC(kernel= "linear", C = C)
    plot_svm_margins(classifier, X, y)
```
